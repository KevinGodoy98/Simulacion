<h2>1 Con los datos de Fifa, organizar a los jugadores de acuerdo al peso en las siguientes escalas y generar un cuadro tipo PIE<h2>

* Debajo 125 Lbs.
* 125-150
* 150-175
* 175 o superior

```python
#importar las librerias necesarias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import io
from google.colab import files
import plotly.graph_objects as go
```

```python
archivo_fifa = files.upload()
```

```python
#Trabajar con otro tipo de datos Fifa
fifa = pd.read_csv(io.StringIO(archivo_fifa['fifa_datos.csv'].decode('utf-8')))
fifa.head(18207)
```

```python
#Tomamos las columnas de interes en este caso -> jugador y peso
df = pd.DataFrame(fifa, columns = ['Name','Weight'])
df.head(18207)
```

```python
#sacamos todos los jugadores menores a 125 lb
peso_menor_125= df[df["Weight"] < 125.0]
#sacamos todos los jugadores entre 125 pero menores a 150 lb
peso_entre_125_150= df[(df['Weight'] >= 125.0) & (df['Weight'] < 150.0)]
#sacamos todos los jugadores entre 150 pero menores a 175 lb
peso_entre_150_175 = df[(df['Weight'] >= 150.0) & (df['Weight'] < 175.0)]
#sacamos todos los jugadores mayores a 175 lb
peso_mayor_175 = df[df["Weight"] > 175.0]
```

```python
#convertimos los pesos a enteros
dato1 = list(map(int, peso_menor_125['Weight']))
dato2 = list(map(int, peso_entre_125_150['Weight']))
dato3 = list(map(int, peso_entre_150_175['Weight']))
dato4 = list(map(int, peso_mayor_175['Weight']))
```

```python
#valor1= (len(dato1)* 1)/ len(df))
valor1= (len(dato1)*100)/len(df)
valor2= (len(dato2)*100)/len(df)
valor3= (len(dato3)*100)/len(df)
valor4= (len(dato4)*100)/len(df)
```

```python
pesos = [valor1,valor2,valor3,valor4]
nombres = ['pesos menores a 125 lbs','pesos entre 125 y 149 lbs','pesos entre 150 y 174 lbs', 'pesos mayores o iguales a 175 lbs']
plt.pie(pesos, labels= nombres, autopct="%0.1f %%")
plt.axis("equal")
plt.title('Pesos')
plt.show()
```

<h3>2.Generar un grafico de barras o histograma de acuerdo a su habilidad (Overall)  en base a los siguientes segmentos contando el número de jugadores<h3>

40
50
60
70

90

```python
overall = pd.DataFrame(fifa, columns = ['Name','Overall'])
overall.head(18207)
```

```python
#Sacamos los primeros 40,50,60,70,80 y 90 jugadores respectivamente
primero= overall.head(40)
segundo= overall.head(50)
tercero= overall.head(60)
cuarto= overall.head(70)
quinto= overall.head(80)
sexto= overall.head(90)
```

```python
converts = list(map(int, primero['Overall']))
converts1 = list(map(int, segundo['Overall']))
converts2 = list(map(int, tercero['Overall']))
converts3 = list(map(int, cuarto['Overall']))
converts4 = list(map(int, quinto['Overall']))
converts5 = list(map(int, sexto['Overall']))
```

```python
#sacamos un promedio mediante : (x1+xn)/n
primeros40 = sum(converts)/40
primeros50 = sum(converts1)/50
primeros60 = sum(converts2)/60
primeros70 = sum(converts3)/70
primeros80 = sum(converts4)/80
primeros90 = sum(converts5)/90

primeros40 = round(primeros40, 3)
primeros50 = round(primeros50, 3)
primeros60 = round(primeros60, 3)
primeros70 = round(primeros70, 3)
primeros80 = round(primeros80, 3)
primeros90 = round(primeros90, 3)
print(primeros40,primeros50,primeros60,primeros70,primeros80,primeros90)
```

```python
height = ['40','50','60','70','80','90']
bars = [primeros40,primeros50,primeros60,primeros70,primeros80,primeros90]
y_pos = np.arange(len(bars))
 
# Create horizontal bars
plt.barh(y_pos, height)
 
# Create names on the x-axis
plt.yticks(y_pos, bars)
 
## Legenda en el eje y
plt.ylabel('Overall X̅')
 
## Legenda en el eje x
plt.xlabel('Numero Jugadores')
 
## Título de Gráfica
plt.title('Promedio overall en base a los Jugadores')
# Show graphic
plt.show()
```
