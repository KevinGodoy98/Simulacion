```python
#%%
# <p>Se tiene 2 dados al lanzarlos y sumo las caras en un rango de 2 a 12
# Generar 2 numeros pseudo aleatorias y se suma las caras lo iteramos en 100,1000,10000 y
# mostrar en una barra de histogramas el valor que mas se repite</p>

import random
from statistics import mode
import matplotlib.pyplot as plot


def dado(n_iteracion):
    sumas = []
    for i in range(0, n_iteracion):
        # print('-----------------iteracion:', i, '-----------------')
        dadoA = random.randint(1, 6)
        # print('Dado A:', dadoA)
        dadoB = random.randint(1, 6)
        # print('Dado B:', dadoB)
        suma = int(dadoA) + int(dadoB)
        sumas.append(suma)
        # print('el resultado es:', suma)

    print('los valores de las sumas son :', sumas)
    valor_mas_repetido = mode(sumas)
    print('El valor mas repetido es:', valor_mas_repetido)
    plot.hist(x=sumas, bins=valor_mas_repetido, color='#F2AB6D', rwidth=0.85)
    plot.title('Histograma'+' '+str(n_iteracion)+' '+'interaciones')
    plot.xlabel('Valores')
    plot.ylabel('Frecuencia')
    plot.show()
```

```python
print('Con 100 iteraciones')
dado(100)
```

```python
print('Con 1000 iteraciones')
dado(1000)
```

```python
print('Con 10000 iteraciones')
dado(10000)
```
